```python
#list comprehension->快速建立list
[f'{num}號' for num in range(1,51)]
```

```python
import numpy as np
import pandas as pd
array_2d = np.random.randint(50,100,size=(50,5))
df1 = pd.DataFrame(array_2d,
             columns=['國文','英文','數學','地理','社會'],
             index = [f'{num}號' for num in range(1,51)]
             )
df1
```

```python
#從dataFrame取出ndarray
df1.values
```

```python
#dataFrame[欄位名稱] subscript
df1['國文'], df1['英文']
```

```python
#Series的選取,使用subscript
#Series[數值的index]
df1['國文'][0]
#Series[數值的slice]
df1['國文'][:5]
#Series[字串的index]
df1['國文']['1號']
#Series[[字串,字串,字串]]
df1['國文'][['1號','2號','3號']]
```

```python
#dataFrame的subscripts
#dataFrame[欄位名稱]
df1['國文']
```

```python
#dataFrame的subscripts
#dataFrame[[欄位名稱,欄位名稱]]
df1[['國文','英文']]
```
